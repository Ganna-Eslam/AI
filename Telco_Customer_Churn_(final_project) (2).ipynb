```python
for col in numerical_cols:
  q1 = data[col].quantile(0.25)
  q3 = data[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr

  data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import RandomizedSearchCV

def build_model(hidden1_units=128, hidden2_units=64, hidden3_units=32, learning_rate=0.001):
    model = Sequential([
        Dense(hidden1_units, activation='relu', input_shape=(x_train.shape[1],)),
        Dense(hidden2_units, activation='relu'),
        Dense(hidden3_units, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=RMSprop(learning_rate=learning_rate),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

param_dist = {
    'hidden1_units': [64, 128, 256],
    'hidden2_units': [32, 64, 128],
    'hidden3_units': [16, 32, 64],
    'learning_rate': [0.0001, 0.001, 0.01],
    'batch_size': [16, 32, 64],
    'epochs': [20, 30, 40]
}

model = KerasClassifier(model=build_model, verbose=0)

random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_dist,
    n_iter=10,
    cv=3,
    verbose=2,
    n_jobs=1
)

random_search_result = random_search.fit(x_train_resampled, y_train_resampled)

print("Best Parameters:", random_search_result.best_params_)

best_model = random_search_result.best_estimator_

y_pred = (y_pred_prob > 0.5).astype(int)

print("Predictions:", y_pred)
```
